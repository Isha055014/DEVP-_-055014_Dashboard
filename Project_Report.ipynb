**DESCRIPTION OF DATA**

**1. Data Source & Size:**

* Source: The dataset for this project was obtained from Kaggle.

**3. Data Dimension:** _(after taking random sample from the dataset)_

* Number of Variables: 16

* Number of Observations: 3001

**4. Data Variable Type:**

* Text Variables: Transaction_ID, Country, Product, Import_Export, Date, Category, Port, Shipping_Method, Supplier, Customer, Payment_Terms

* Number Variables: Integer: Quantity, Customs_Code, Invoice_Number

* Decimal: Value, Weight

**5. Data Variable Category:**

* Index: Transaction_ID, Invoice_Number (used for identifying each transaction)

* Categorical (Nominal): Country, Product, Import_Export, Category, Port, Shipping_Method, Supplier, Customer, Payment_Terms

* Categorical (Ordinal):

* Non-Categorical: Quantity, Value, Customs_Code, Weight

**ANALYSIS AND MANAGERIAL INSIGHTS**

* **1. AVERAGE VALUE** 

  * **ANAYLSIS**: 
  After comparing, managers can observe the major (mean) valued item to be exported or imported and can increase the maunfactuirng of the same to increase the profit.
  To avoid reliance on a few high-value goods, it might be beneficial to explore introducing mid-range products that increase the total volume without significantly dropping the average value.

  * **MANAGERIAL INSIGHTS**: Monitoring the average value of imports helps managers understand the demand for foreign goods. If this value fluctuates significantly over time, managers might explore reasons such as global market demand, product mix changes, or economic conditions.

* **2. AVERAGE WEIGHT**

  * **ANAYLSIS**: Managers should analyze the relationship between the average weight and average value to ensure that the cost of shipping isn't disproportionately high compared to the revenue generated.
  Also they can analyse which shipping method does not require heavy packing which requires heavy costs.

  * **MANAGERIAL INSIGHTS**: By comparing the value-weight ratios managers can focus on sales, marketing and manufacuring of the most demanded category (categories) and can reduce their packing costs.
  Managers could also consider bulk shipping or negotiating better freight rates for consistent weights.


* **3. AVERAGE QUANTITY**: 

  * **ANAYLSIS**: This can prompt strategic decisions about product discontinuation or revitalization efforts.

  * **MANAGERIAL INSIGHTS**: Monitoring average quantities can indicate where products are in their life cycles. A decline in average quantity might suggest that a product is nearing the end of its lifecycle.



* **GENERAL INSIGHTS**:

  * Managers should balance between the trade value, weight, and quantity. For example, if the weight is high but the value is low, it may be wise to reconsider shipping methods or product focus.

  * These averages can be used to segment customers based on order size, allowing for tailored marketing and pricing strategies.

  * Reviewing this data could help streamline operations such as warehouse management, shipping, and customs clearance based on typical order size and weight.

**ANALYSIS**:

  * The min value of air (1.64) is slightly higher than sea and land, which might indicate that air shipping is avoided for very lightweight items due to higher costs associated with air freight.

  * Land shipping has a slightly higher median weight compared to sea, suggesting that it is used more for moderately heavy goods but remains versatile for lower weights as well.

  * Sea shipping has a broad range of weights, with a low minimum of 0.57 and a median that is slightly lower than air and land. This reflects that sea freight is used for a wide variety of goods, including both lightweight and bulk shipments.

**MANAGERIAL INSIGHTS**:

   '< 500 kg: Air freight (if urgent), else land.
   '500 - 2000 kg: land or sea depending distance.
   '> 2000 kg: Sea freight for international, land for domestic/regional.

  * They can also identify customers who prioritize speed over cost (likely air freight customers) and those who prefer lower costs for bulk orders (sea freight). This allows for targeted pricing strategies and marketing campaigns that appeal to different customer needs.
