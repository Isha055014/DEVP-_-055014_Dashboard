**1. Data Source & Size:**

* Source: The dataset for this project was obtained from Kaggle.

* Size: 15,000 records with 16 variables.

**2. Data Type:** 
* This is a Cross-sectional dataset, capturing transactional data at a specific point in time.

**3. Data Dimension:**

* Number of Variables: 16

* Number of Observations: 15,000

**4. Data Variable Type:**

* Text Variables: Transaction_ID, Country, Product, Import_Export, Date, Category, Port, Shipping_Method, Supplier, Customer, Payment_Terms

* Number Variables: Integer: Quantity, Customs_Code, Invoice_Number

* Decimal: Value, Weight

**5. Data Variable Category:**

* Index: Transaction_ID, Invoice_Number (used for identifying each transaction)

* Categorical (Nominal): Country, Product, Import_Export, Category, Port, Shipping_Method, Supplier, Customer, Payment_Terms

* Categorical (Ordinal):

* Non-Categorical: Quantity, Value, Customs_Code, Weight

  * **MANAGERIAL INSIGHTS**: Managers should adjust production focus, marketing, or sales efforts based on weight trends of different product categories.


* **3. AVERAGE QUANTITY**: 

  * **ANAYLSIS**: This can prompt strategic decisions about product discontinuation or revitalization efforts.

  * **MANAGERIAL INSIGHTS**: Monitoring average quantities can indicate where products are in their life cycles. A decline in average quantity might suggest that a product is nearing the end of its lifecycle.
